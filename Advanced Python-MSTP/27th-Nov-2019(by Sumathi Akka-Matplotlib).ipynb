```python
a=np.arange(10)
plt.plot(a,a*2,'r')
plt.plot(a,a**2,'b')
plt.plot(a,a*a+2*a,'go--')
```

```python
plt.figure()
plt.plot(a,a*3+2,'go-')
plt.title('Line Equation')
plt.xlabel('a values')
plt.ylabel('a*3+2 values')
```

```python
plt.figure(figsize=(6,3)) # size of the graph(width,Height)
plt.plot(a,a*3+2,'g')
plt.plot(x,y,'r')
plt.title('Line Equation')
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
plt.figure(figsize=(6,3)) # size of the graph(width,Height)
plt.plot(a,a*3+2,'g')
plt.xlim(5,8)
plt.ylim(15)
```

Subplots
- only plt functions
- Combination of both sns and plt

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Line')
plt.subplot(2,1,2)
plt.plot(x,x*x,'go')
plt.title('Subplot')
plt.title('Y=X*2')
plt.tight_layout()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Y=X*2')
plt.subplot(2,1,2)
plt.plot(x,x*x,'go')
plt.subplot(1,2,1)
plt.plot(x,y)
```

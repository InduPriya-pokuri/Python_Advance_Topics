```python
import pandas as pd

df = pd.read_csv('olympics.csv',index_col=0,skiprows=1)
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]},inplace=True) # removing unwanted columns
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[2:]},inplace=True) # renaming 02 with silver
    if col[:3]=='03':
        df.rename(columns={col:'Bronze'+col[2:]},inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]},inplace=True)
df
```

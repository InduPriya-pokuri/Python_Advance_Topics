```python
cust_df =pd.read_csv('customer.csv')
prod_df =pd.read_csv('product.csv')
ship_df =pd.read_csv('shipping.csv')
order_df=pd.read_csv('orders.csv')
```

```python
df1=pd.merge(df,cust_df,on='Cust_id')
df1
```

```python
df1.shape
```

```python
#Merge the df1 and prod_df files and store the result  in df2
df2=pd.merge(df1,prod_df,on='Prod_id')
df2
```

```python
#Merge the df2 and ship_df files and store the result  in df3
df3=pd.merge(df2,ship_df,on='Ship_id')
df3
```

```python
df3.shape
```

```python
#Merge the df3 and order_df files and store the result  in master_df
master_df=pd.merge(df3,order_df,on='Ord_id')
master_df
```
